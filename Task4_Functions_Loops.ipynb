```python
# Problem 1: Mobile Recharge Validation System

def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("‚ùå Minimum recharge amount is ‚Çπ50")
        return False
    elif amount not in valid_plans:
        print("‚ùå Invalid plan selected")
        return False
    else:
        print("‚úÖ Recharge Successful!")
        return True


# Retry using while loop
while True:
    try:
        amount = int(input("Enter recharge amount: "))
        if validate_recharge(amount):
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Enter a valid number.\n")
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock = {stock})")
        else:
            print(f"{product}: Stock OK (Stock = {stock})")


# Sample dictionary
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 25,
    "Soap": 12
}

check_inventory(inventory)
```

```python
# Problem 3: Student Result Processing System

def process_result(marks):
    total = 0

    # Loop to calculate total
    for mark in marks:
        total += mark

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Sample input
marks_list = [60, 70, 45, 80, 55]
result = process_result(marks_list)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator

def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    total_fare = base_fare + (distance * per_km)

    if peak.lower() == "yes":
        total_fare += total_fare * 0.25   # 25% extra

    return total_fare


# Retry using while loop
while True:
    try:
        distance = float(input("Enter distance (km): "))
        peak = input("Is it peak hour? (yes/no): ")

        fare = calculate_fare(distance, peak)
        print("Total Fare: ‚Çπ", fare)

        break
    except ValueError:
        print("Invalid input. Please try again.\n")
```

```python
# Problem 5: Attendance Eligibility Checker

def check_attendance(attendance_list):
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage, "%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

```python
# Problem 6: Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "‚ùå Weak Password (Minimum 8 characters required)"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "‚úÖ Strong Password"
    else:
        return "‚ùå Weak Password (Must contain digit and special character)"


# Input
password = input("Enter password: ")
print(check_password(password))
```
