```python
# Correct credentials
correct_username = "admin"
correct_password = "1234"

# User input
entered_username = input("Enter username: ")
entered_password = input("Enter password: ")

# Checking credentials
if entered_username == correct_username and entered_password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Loop through each mark
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Display results
print("Total Students:", len(marks))
print("Number of Passed Students:", pass_count)
print("Number of Failed Students:", fail_count)
```

```python
# Original list
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    cleaned_name = name.strip().lower()   # Remove spaces & convert to lowercase
    cleaned_names.append(cleaned_name)

# Display cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    message_length = len(message)
    print("Message:", message)
    print("Length:", message_length)

    # Flag messages longer than 10 characters
    if message_length > 10:
        print("Status: Long Message")

    print()  # Blank line for readability
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR messages
for log in logs:
    if log == "ERROR":
        error_count += 1

# Display result
print("Total ERROR Messages:", error_count)
```
